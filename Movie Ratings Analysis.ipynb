```python
# Co-ordinate and Diagonals
```

```python
FG = sns.FacetGrid(movies,row='Genre',col='Year_Release',hue='Genre')
kws= dict(s=50, linewidth=0.5, edgecolor='black')
FG = FG.map(plt.scatter,'CriticRating','AudianceRating', **kws)
FG.set(xlim=(0,100), ylim=(0,100))
for ax in FG.axes.flat:
    ax.plot((0,100), (0,100), ls='--')
FG.add_legend()
```
